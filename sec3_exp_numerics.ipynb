# About

This script generates the numerical plots for an N-dim Kuramoto system with a constant, non-adaptive delay (exponential) distribution), with repeated trials under a fixed history function. The purpose is show the relation between the mean delay and the global coefficient $g$.

All trials are performed in MATLAB. The MATLAB script is found in 'sec3_dirac_numerics.mat'

From folder, import the 'parameters.mat' file. Here, 'g', 'taum' are arrays that index the solution files in the form '{index_g}_{index_taum}.mat'.

# Library

```python
from __future__ import division, print_function

%matplotlib inline
# Toggle on/off
# %matplotlib notebook

import os
import numpy as np
import scipy.io as sio
from scipy import optimize
import matplotlib.pyplot as plt
import matplotlib.colors as clr
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.transforms as tsfm
import math
from math import pi

from lib import *

from IPython.display import clear_output
```

# Directories

```python
# Data files from Matlab
import_folder = 'sec3_exp_numerics'

# Directory
dir_main = os.getcwd()
dir_im = os.path.join(dir_main, 'matlab', 'data', import_folder)

filenames = os.listdir(dir_im)
filenames.remove('parameters.mat')
```

# Import and compile data

## Parameters

```python
dir_par = os.path.join(dir_im, 'parameters.mat')
dict_par = sio.loadmat(dir_par)

taum_arr = np.reshape(dict_par['taum'], -1)

g = dict_par['g'][0,0]
w0 = dict_par['omega0'][0,0]
t0 = dict_par['t0'][0,0]
tf = dict_par['tf'][0,0]
Omega0 = dict_par['Omega0'][0,0]
std = dict_par['std'][0,0]
N = dict_par['N'][0,0]

# Asymptotic percentage
asy = 0.1
```

## Import data

```python
# Compile arrays
asyOrders = np.zeros(len(filenames))
asyOmegas = np.zeros(len(filenames))
eqOmegas = np.zeros(len(filenames))
maxReEigs = np.zeros(len(filenames))
arr_taum = np.zeros(len(filenames))

indPlot1 = 4
indPlot2 = 12

zeroTol = 0.01

# Acquire trials data
for i in range(len(filenames)):
    
    # Import .mat file
    filename = filenames[i]
    dir_mat = os.path.join(dir_im, filename)
    dict_mat =  sio.loadmat(dir_mat)
    
    # Get g, tau0 indices
    ind_file = filename.split('.')[0]
    ind = int(ind_file)
    taum = taum_arr[ind]
    
    # Compile solution dictionary
    sol = {'t': np.reshape(dict_mat['t'], -1),
           'y': dict_mat['y'],
           'yp': dict_mat['yp'],
           'tau': dict_mat['tau']
          }
    
    # Compute the global frequency:
    arrOmega = asylib.weight_avg(sol['t'], sol['yp'], asy)
    asyOmega = np.sum(arrOmega) / arrOmega.size
    
    # Get array of phases
    arrLinear = asyOmega*sol['t']
    arrLinear = arrLinear[:,None]

    arrPhases = sol['y'] - arrLinear
    sol['phases'] = arrPhases
    
    # Order parameter
    orderY = funlib.order(sol['y'])
    sol['order'] = orderY.copy()
    orderY = orderY.reshape(-1,1)
    asyOrder = np.sum(asylib.weight_avg(sol['t'], orderY, asy))
    
    # Sync roots
    # roots = dict_mat['eqOmegas']
    # eqOmega = roots[np.isreal(roots)][0].real
    eqOmega = dict_mat['eqOmega'][0,0]
    
    # Maximum non-zero eigenvalue
    eigs = dict_mat['found']
    eigs = eigs[np.abs(eigs) > zeroTol]
    if eigs.size == 0:
        maxEig = 0
    else:
        maxEig = np.max(eigs.real)
    
    # Filter eigenvalues with large log10 error
    eigs = eigs[]
    
    # Store
    asyOrders[ind] = asyOrder
    asyOmegas[ind] = asyOmega
    arr_taum[ind] = taum
    eqOmegas[ind] = eqOmega
    maxReEigs[ind] = maxEig
    
    # Store solutions
    if ind == indPlot1:
        sol1 = sol.copy()
        sol1['order'] = orderY
        sol1['phases'] = arrPhases
        sol1['taum'] = taum
        sol1['tau'] = dict_mat['tau']
        sol1['maxeig'] = maxEig
        
    elif ind == indPlot2:
        sol2 = sol.copy()
        sol2['order'] = orderY
        sol2['phases'] = arrPhases
        sol2['taum'] = taum
        sol2['tau'] = dict_mat['tau']
        sol2['maxeig'] = maxEig
        
# Sort with increasing taum
inds_sort = np.argsort(arr_taum)
arr_taum = arr_taum[inds_sort]
asyOrders = asyOrders[inds_sort]
asyOmegas = asyOmegas[inds_sort]
maxReEigs = maxReEigs[inds_sort]
eqOmegas = eqOmegas[inds_sort]
eqOmegas[0] = w0
```

## Histogram

```python
tau1 = np.reshape(sol1['tau'], -1)
tau2 = np.reshape(sol2['tau'], -1)

# Bins
bin_range = (0, 10)
num_bins = 20
```

# Plot figure 1 (individual trials)

## Options

```python
# Plot positions [left, bottom, width, height]
bbox0 = [0.05, 0.65, 0.20, 0.35] # Top left (trial 1 freqs)
bbox1 = [0.05, 0.10, 0.20, 0.35] # Bottom left (trial 1 phases)
bbox2 = [0.30, 0.65, 0.20, 0.35] # Top right (trial 2 freqs)
bbox3 = [0.30, 0.10, 0.20, 0.35] # Bottom right (trial 2 phases)
bbox4 = [0.62, 0.10, 0.38, 0.35] # Right (order)
bbox5 = [0.62, 0.60, 0.38, 0.40] # Right (order)

# Alphas
alpha1 = 0.8
alpha_col = 0.2
alpha_w0 = 0.8

# Colours
cmap1 = plt.get_cmap('spring')
cmap2 = plt.get_cmap('cool')

num1 = sol1['yp'].shape[1]
num2 = sol2['yp'].shape[1]

colors1 = [cmap1(i) for i in np.linspace(0, 0.99, num1)]
colors2 = [cmap2(i) for i in np.linspace(0, 0.99, num2)]

col1 = 'orange'
col2 = 'blue'
col_w0 = 'green'

col1_alp = clr.to_rgba(col1, alpha=alpha_col)
col2_alp = clr.to_rgba(col2, alpha=alpha_col)

# Linewidths
lw1 = 0.5
lw2 = 1.0
```

## Generate figure

```python
fig, ax = plt.subplots(6, figsize=(12,4), dpi=80)

# Set positions
ax[0].set_position(bbox0)
ax[1].set_position(bbox1)
ax[2].set_position(bbox2)
ax[3].set_position(bbox3)
ax[4].set_position(bbox4)
ax[5].set_position(bbox5)


# Plot trial 1:
ax[0].set_prop_cycle(color=colors1)
ax[1].set_prop_cycle(color=colors1)
ax[0].plot(sol1['t'], sol1['yp'], alpha=alpha1, linewidth=lw1, zorder=2)
ax[1].plot(sol1['t'], np.sin(sol1['phases']), alpha=alpha1, linewidth=lw1, zorder=2)

# Plot trial 2:
ax[2].set_prop_cycle(color=colors2)
ax[3].set_prop_cycle(color=colors2)
ax[2].plot(sol2['t'], sol2['yp'], alpha=alpha1, linewidth=lw1, zorder=2)
ax[3].plot(sol2['t'], np.sin(sol2['phases']), alpha=alpha1, linewidth=lw1, zorder=2)

# Orders
ax[4].plot(sol1['t'], sol1['order'], color=col1, alpha=alpha1, zorder=2)
ax[4].plot(sol2['t'], sol2['order'], color=col2, alpha=alpha1, zorder=2)

# Histogram
ax[5].hist(tau1, bins=num_bins, range=bin_range, density=True, color=col1_alp, edgecolor=col1, linewidth=lw2)
ax[5].hist(tau2, bins=num_bins, range=bin_range, density=True, color=col2_alp, edgecolor=col2, linewidth=lw2)

# Adjust axes
for i in range(len(ax)):
    ax[i].set_xlim(left=t0, right=tf/5)

ax[5].set_xlim(left=bin_range[0], right=bin_range[1])

# Axis lines
ax[0].axhline(y=w0, color=col_w0, zorder=0)
ax[2].axhline(y=w0, color=col_w0, zorder=0)
```
