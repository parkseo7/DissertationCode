# About

This script generates the numerical plots for the 2-dim Kuramoto system with adaptive delays, with varying initial points.
The purpose is to show the numerical basin of attraction around each stable synchronous point. All trials are performed in MATLAB. The MATLAB script is found in 'sec4_2D_numerics2.mat'

# Import and compile data

## Numerical arrays and asymptotic values

```python
def process_folder(directory, inds=[]):
    '''
    Function to compile folder given by directory into processed meshes. Here, (eqOmegas, eqDeltas) are the theoretical
    synchronization frequencies and phase differences. threshold is the proximity for which the points must be in to be included
    as convergent to the theoretical point. decimals is the number of relevant decimals in the frequencies and phases.
    '''
    
    # Get filenames
    filenames = os.listdir(directory)
    filenames.remove('parameters.mat')
    
    # Get parameters
    dir_par = os.path.join(directory, 'parameters.mat')
    dict_par = sio.loadmat(dir_par)

    parameters = {'g': dict_par['g'][0,0]/2,
                  'omega0': dict_par['omega0'][0,0],
                  'gain': dict_par['gain'][0,0],
                  't0': dict_par['t0'][0,0],
                  'tf': dict_par['tf'][0,0],
                  'tau0': dict_par['tau0'][0,0],
                  'alphatau': dict_par['alphatau'][0,0],
                  'asy': dict_par['asy'][0,0]
                 }
    
    # Initial values
    Omega0_arr = np.reshape(dict_par['Omega0_arr'], -1)
    Delta0_arr = np.reshape(dict_par['Delta0_arr'], -1)
    
    # Obtain all theoretical asymptotic values
    Omegafun = funlib.Omega2D(parameters)
    
    w0 = parameters['omega0']
    g = parameters['g']
    
    # Get all roots of Omega_fun in region
    eq_Omegas = rootlib.root_on_interval(Omegafun, w0-g, w0+g, steps=1000)
    eq_Deltas = np.arcsin((w0 - eq_Omegas)/g)

    # Compiled arrays
    Omega_mesh, Delta_mesh = np.meshgrid(Omega0_arr, Delta0_arr)
    asyOmega_mesh = np.empty(Omega_mesh.shape)
    asyOmega_mesh[:] = np.NaN
    varDelta_mesh = np.zeros(Omega_mesh.shape)
    
    # Indices to save full solution arrays
    sols = {}
        
    # COMPILE DATA
    for i in range(len(filenames)):
    
        # Import .mat file
        filename = filenames[i]
        dir_mat = os.path.join(directory, filename)
        dict_mat =  sio.loadmat(dir_mat)

        # Get g, tau0 indices
        inds_file = filename.split('.')[0]
        inds_file = inds_file.split('_')
        ind = (int(inds_file[0]), int(inds_file[1]))
    
        # Compile solution dictionary
        sol = {'t': np.reshape(dict_mat['t'], -1),
               'y': dict_mat['y'],
               'yp': dict_mat['yp'],
               'tau': dict_mat['tau'],
               'taup': dict_mat['taup'],
               'histt': dict_mat['histt'],
               'histy': dict_mat['histy'],
               'histyp': dict_mat['histyp']
              }
          
        init_Omega = dict_mat['Omega0'][0,0]
        init_Delta = dict_mat['Delta0'][0,0]
        
        # Asymptotic Omega:
        asy_Omega = dict_mat['Omega_asy'][0,0]
        asy_Delta = dict_mat['Delta_asy'][0,0]
        asy_phases = dict_mat['phases_asy']
        
        arrlinear = asy_Omega*sol['t']
        arrlinear = arrlinear[:,None]
        arrphases = sol['y'] - arrlinear
        
        histlinear = asy_Omega*sol['histt']
        histphases = sol['histy'] - histlinear

        # Variance
        var_Delta = np.average(asylib.weight_avg(sol['t'], (arrphases - asy_phases)**2, parameters['asy']))
        
        # Save to arrays
        asyOmega_mesh[ind] = asy_Omega
        varDelta_mesh[ind] = var_Delta
        
        if ind in inds:
            sols[ind] = sol.copy()
            sols[ind]['arrphases'] = arrphases
            sols[ind]['histphases'] = histphases
            sols[ind]['initOmega'] = init_Omega
            sols[ind]['initDelta'] = init_Delta
            sols[ind]['asyOmega'] = asy_Omega
            sols[ind]['asyDelta'] = asy_Delta
    
    # Compile as dictionary
    output = {'Omega_mesh': Omega_mesh,
              'Delta_mesh': Delta_mesh,
              'asyOmega_mesh': asyOmega_mesh,
              'varDelta_mesh': varDelta_mesh,
              'sol': sols,
              'parameters': parameters,
              'eqOmegas': eq_Omegas,
              'eqDeltas': eq_Deltas
              }
    
    return output
```

# Figure 2 (asymptotic)

## Configure arrays

```python
def configure(trial, threshold=0.1):
    '''
    Given trial dictionary (output from the 'process_folder' function), returns a dictionary of heatmap arrays.
    Threshold is the maximum distance |Omega - Omega0| allowed for synchrony to be considered.
    '''
    
    # Acquire meshes
    Omega_mesh = trial['Omega_mesh']
    Delta_mesh = trial['Delta_mesh']
    asy_mesh = trial['asyOmega_mesh']
    
    # Parameters
    parameters = trial['parameters']
    
    # Theoretical sync states:
    eqOmegas = trial['eqOmegas']
    eqDeltas = trial['eqDeltas']
    
    # Index mesh
    ind_mesh = np.zeros(Omega_mesh.shape, dtype='int64')
    
    dim = ind_mesh.shape
    for i in range(dim[0]):
        for j in range(dim[1]):
            asy_Omega = asy_mesh[i,j]
            minDist = np.min(np.abs(asy_Omega - eqOmegas))
            minind = np.argmin(np.abs(asy_Omega - eqOmegas))
            
            if minDist < threshold:
                ind_mesh[i,j] = minind+1    
                
    
    # Compile necessary arrays
    arrs = {'Omega': Omega_mesh,
            'Delta': Delta_mesh,
            'asyOmega': asy_mesh,
            'asyinds': ind_mesh,
            'eqOmega': eqOmegas,
            'eqDelta': eqDeltas,
            'parameters': parameters
            }
    
    return arrs


# Configure
P1 = configure(tr1, threshold=0.01)
P2 = configure(tr2, threshold=0.01)
P3 = configure(tr3, threshold=0.01)
```

## Figure procedure

Apply the function below to each compiled mesh.

```python
def plot_basin(ax0, plot_arrs0):
    '''
    Function to plot the meshes as a colormap.
    '''
    
    # Get plot arrays
    plot_arrs = plot_arrs0.copy()
    Omega_mesh = plot_arrs['Omega']
    Delta_mesh = plot_arrs['Delta']
    ind_mesh = plot_arrs['asyinds']
    eq_Omegas = plot_arrs['eqOmega']
    eq_Deltas = plot_arrs['eqDelta']
    parameters = plot_arrs['parameters']
    
    # OPTIONS
    
    # Colour map
    cmapname = 'YlOrRd'
    cmap = plt.get_cmap(cmapname)

    # Markers
    markersize = 200
    markersizegrid = 12
    marker = '*'

    # Lines
    linecolor = 'green'
    linewidth = 1.5

    # Alphas
    alphaline = 0.8
    alphamap = 0.5
    
    # Major ticks
    xticks = Delta_mesh[:,0]
    yticks = Omega_mesh[0,:]
    majorevery = 2
    
    # Maximum index
    INDMAX = np.int64(np.max(ind_mesh))
    
    # Plot colormap
    inds_level = np.sort(np.unique(ind_mesh))
    if inds_level.size > 1:
        levels = np.linspace(inds_level[0], inds_level[-1], inds_level.size+1, endpoint=True)
    else:
        levels = np.linspace(inds_level[0]-1, inds_level[0], 3, endpoint=True)
    print(levels)
    ax0.contourf(Delta_mesh, Omega_mesh, ind_mesh, levels, cmap=cmapname, alpha=alphamap)
    
    ax0.set_xticks(xticks[np.arange(0, xticks.size, majorevery)])
    ax0.set_xticks(xticks, minor=True)
    ax0.set_yticks(yticks[np.arange(0, yticks.size, majorevery)])
    ax0.set_yticks(yticks, minor=True)
    
    ax0.grid(b=True, which='both', color='black', alpha=0.4, zorder=-1)
    
    # Plot state markers
    for k in range(inds_level.size):
        ind0 = inds_level[k]
        if inds_level.size == 1:
            colfrac = 1.0
        else:
            colfrac = (ind0 - inds_level[0]) / (inds_level[-1] - inds_level[0])
        markcol = cmap(colfrac)
        # markseq = np.linspace(0, 1, num=INDMAX+2, endpoint=True)[1:]
        # markcols = cmap(markseq)
        ax0.scatter(eq_Deltas[ind0-1], eq_Omegas[ind0-1], c=markcol, marker=marker, s=markersize, zorder=3)
    
    # Plot data markers
    # markcolsgrid = cmap(ind_mesh.ravel()/INDMAX)
    # ax0.scatter(Delta_mesh.ravel(), Omega_mesh.ravel(), c=markcolsgrid, s=markersizegrid, zorder=2)
    
    # Set limits
    ax0.set_xlim(left=np.min(Delta_mesh), right=np.max(Delta_mesh))
    ax0.set_ylim(bottom=np.min(Omega_mesh), top=np.max(Omega_mesh))
    
    # Horizontal lines
    ax0.axhline(y=parameters['omega0'], color=linecolor, linewidth=linewidth, zorder=1)
```

## Generate figure

```python
fig, ax = plt.subplots(3, figsize=(12,4), dpi=400)

# Set positions
ax[0].set_position(Bbox0)
ax[1].set_position(Bbox1)
ax[2].set_position(Bbox2)

plot_basin(ax[0], P1)
plot_basin(ax[1], P2)
plot_basin(ax[2], P3)
```
