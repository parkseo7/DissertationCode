# About

Script to generate the figures pertaining to large-N dimensional analysis. Includes plotting the synchronous frequencies and Gaussian variance and their corresponding eigenvalue distributions and patterns. Import .mat files generated from 'sec4_ND_analysis.m'.

# Directories

```python
# Data files (for fig.4d)
import_folder = 'sec4_ND_analysis'

# Directory
dir_main = os.getcwd()
dir_im = os.path.join(dir_main, 'matlab', 'data', import_folder)

filenames = os.listdir(dir_im)
filenames.remove('parameters.mat')
```

# Import

## Parameters

```python
dir_sam = os.path.join(dir_im, 'parameters.mat')
dict_mat = sio.loadmat(dir_sam)

# Parameters
deltaarr = np.reshape(dict_mat['delta_arr'], -1)
Omegaarr = np.reshape(dict_mat['Omega_arr'], -1)
g = dict_mat['g'][0,0]
w0 = dict_mat['omega0'][0,0]
gain = dict_mat['gain'][0,0]
tau0 = dict_mat['tau0'][0,0]
scale = dict_mat['scale'][0,0]

tol = dict_mat['tol'][0,0]
rerange = np.reshape(dict_mat['rerange'], -1)
imrange = np.reshape(dict_mat['imrange'], -1)
```

## Theoretical frequencies

```python
parameters = {'g': g, 'omega0': w0, 'gain': gain, 'tau0': tau0}
L = 2*pi
steps = 300

# Choose deltas
delta01 = 0.01
delta02 = 0.05
delta03 = 0.1
delta04 = 0.20

freqarr = np.linspace(w0-g, w0+g, num=100)
errarr1 = np.zeros(freqarr.size)
errarr2 = np.zeros(freqarr.size)
errarr3 = np.zeros(freqarr.size)
errarr4 = np.zeros(freqarr.size)

for i in range(len(freqarr)):
    freq = freqarr[i]
    errarr1[i] = freq - funlib.Omega_infty(freq, delta01, parameters, L=L, steps=steps)
    errarr2[i] = freq - funlib.Omega_infty(freq, delta02, parameters, L=L, steps=steps)
    errarr3[i] = freq - funlib.Omega_infty(freq, delta03, parameters, L=L, steps=steps)
    errarr4[i] = freq - funlib.Omega_infty(freq, delta04, parameters, L=L, steps=steps)
```

## Compile arrays

```python
# Chosen deltas
ind1 = 4
ind2 = 6
ind3 = 8

# Tolerance for zero
tolzero = 0.1
threshold = 1

leadReEigs = np.zeros(len(filenames))
leadImEigs = np.zeros(len(filenames))

for i in range(len(filenames)):
    
    # Import .mat file
    filename = filenames[i]
    dir_mat = os.path.join(dir_im, filename)
    dict_mat =  sio.loadmat(dir_mat)
    
    # Get g, tau0 indices
    ind_file = filename.split('.')[0]
    ind = int(ind_file)
    
    # Acquire data
    delta = dict_mat['delta'][0,0]
    Omega = dict_mat['Omega'][0,0]
    
    # Scale
    scale = dict_mat['scale'][0,0]
    
    # Eigenvalues
    eigs = dict_mat['found']
    errors = dict_mat['errors']
    
    # Only keep eigenvalues below threshold:
    eigsnonzero = eigs[np.log10(1+np.abs(errors)) < threshold]
    scaled_eigs = eigsnonzero
    
    # Leading eigenvalue
    nonzero_eigs = scaled_eigs[np.abs(scaled_eigs) > tolzero]
    if nonzero_eigs.size == 0:
        leadreeig = 0
        leadimeig = 0
    else:
        leadreeig = np.max(np.real(nonzero_eigs))
        ind_img = np.argmax(np.real(nonzero_eigs))
        leadimeig = np.imag(nonzero_eigs[ind_img])
    
    # Update arrays
    leadReEigs[i] = leadreeig
    leadImEigs[i] = leadimeig
    
    if ind == ind1:
        u_mesh1 = dict_mat['re']
        v_mesh1 = dict_mat['im']
        f_mesh1 = dict_mat['fvals']
        eigs1 = eigs
        errs1 = errors
        leadReEig1 = leadreeig
        delta1 = delta
        Omega1 = Omega
        
    if ind == ind2:
        u_mesh2 = dict_mat['re']
        v_mesh2 = dict_mat['im']
        f_mesh2 = dict_mat['fvals']
        eigs2 = eigs
        errs2 = errors
        leadReEig2 = leadreeig
        delta2 = delta
        Omega2 = Omega
        
    if ind == ind3:
        u_mesh3 = dict_mat['re']
        v_mesh3 = dict_mat['im']
        f_mesh3 = dict_mat['fvals']
        eigs3 = eigs
        errs3 = errors
        leadReEig3 = leadreeig
        delta3 = delta
        Omega3 = Omega
```

# Generate figure 2 (eigenvalues)

## Eigenvalue plots

Define function to plot each set of meshes and points

```python
def plot_eigenvalues(ax, u_mesh, v_mesh, f_mesh, eigenvalues, cax=None):
    '''
    Plots a heat map using u_mesh, v_mesh, and err_mesh. Plots the eigenvalues within range.
    '''
    
    # OPTIONS
    
    # Colour map
    cmapname = 'binary'
    cmap = plt.get_cmap(cmapname)

    # Markers
    markersize = 50
    marker = 'X'
    markercolor = 'blue'
    
    # Axis lines options
    linewidth = 0.5
    linecolor = 'black'
    
    # Alphas
    alphamap = 0.8
    
    # levels = np.linspace(0, INDMAX, INDMAX+1, endpoint=True)
    logerror = np.log10(1+np.abs(f_mesh - u_mesh - 1j*v_mesh))
    maxerror = np.ceil(np.max(logerror))
    levels = np.linspace(0, maxerror, 40)
    cs = ax.contourf(u_mesh, v_mesh, logerror, levels, cmap=cmapname, alpha=alphamap, zorder=1)
    
    # Plot eigenvalues
    ax.scatter(eigenvalues.real, eigenvalues.imag, color=markercolor, marker=marker, s=markersize, zorder=3)
    
    # Axis lines
    ax.axhline(y = 0, color=linecolor, linewidth=linewidth, zorder=2)
    ax.axvline(x = 0, color=linecolor, linewidth=linewidth, zorder=2)
    
    if cax != None:
        tickinterval = 1
        cbar = fig.colorbar(cs, cax=cax, ticks=np.arange(0, maxerror, tickinterval))
```

## Options

```python
# Plot positions [left, bottom, width, height] 
bbox0 = [0.1, 0.6, 0.2, 0.4] # Fig 2a: Leading real eigenvalue
bbox1 = [0.1, 0.1, 0.2, 0.4] # Fig 2b: Leading imaginary eigenvalue
bbox2 = [0.40, 0.1, 0.15, 0.9] # Fig 2c: Eigenvalue plot 1
bbox3 = [0.60, 0.1, 0.15, 0.9] # Fig 2d: Eigenvalue plot 2
bbox4 = [0.80, 0.1, 0.15, 0.9] # Fig 2e: Eigenvalue plot 3
bbox5 = [0.97, 0.1, 0.03, 0.9] # Fig 2f: Color legend of eigenvalue plots

# Sign log function
slog = lambda x: np.sign(x)*np.log10(1 + np.abs(x))

# Start plotting deltas at index:
i_s = 2
```

```python
fig, ax = plt.subplots(6, figsize=(12,6), dpi=80)

# Set positions
ax[0].set_position(bbox0)
ax[1].set_position(bbox1)
ax[2].set_position(bbox2)
ax[3].set_position(bbox3)
ax[4].set_position(bbox4)
ax[5].set_position(bbox5)

# Plot errors
ax[0].plot(deltaarr[i_s:], leadReEigs[i_s:], marker='o', color='green', markersize=5, zorder=2)
ax[1].plot(deltaarr[i_s:], leadImEigs[i_s:], marker='o', color='magenta', markersize=5, zorder=2)
ax[1].plot(deltaarr[i_s:], -leadImEigs[i_s:], marker='o', color='magenta', markersize=5, zorder=2)

# Edit leading eigenvalue plots
ax[0].set_xlim(left=0, right=np.max(deltaarr))
ax[0].axhline(y=0, linewidth=linewidthA, color=linecolorA, zorder=0)
ax[1].set_xlim(left=0, right=np.max(deltaarr))
ax[1].axhline(y=0, linewidth=linewidthA, color=linecolorA, zorder=0)

# Plot eigenvalues
plot_eigenvalues(ax[2], u_mesh1, v_mesh1, f_mesh1, eigs1)
plot_eigenvalues(ax[3], u_mesh2, v_mesh2, f_mesh2, eigs2)
plot_eigenvalues(ax[4], u_mesh3, v_mesh3, f_mesh3, eigs3, cax=ax[5])
```
